```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets
```

```python
bc=datasets.load_breast_cancer()
x=bc.data
y=bc.target
```

```python
col1=bc.feature_names
col2=bc.target_names
```

```python
df=pd.DataFrame(x,columns=col1)
```

```python
df.clip(df.quantile(0.25),df.quantile(0.75),axis=1,inplace=True)
```

```python
sns.boxplot(df)
```

```python
from sklearn.preprocessing import StandardScaler,Normalizer,MinMaxScaler,RobustScaler
```

```python
S=StandardScaler()
x=S.fit_transform(df)
```

```python
x=pd.DataFrame(x,columns=col1)
```

```python
x.head()
```

```python
df['target']=y
```

```python
df
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(df.corr(),annot=True)
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import mutual_info_classif
```

```python
imp=mutual_info_classif(x,y)
feat_imp=pd.Series(imp,df.columns[0:len(df.columns)-1])
feat_imp.plot(kind='barh',color='Red')
plt.show()
```

```python
a=SelectKBest(mutual_info_classif,k=11)
X=a.fit_transform(x,y)
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
k=KNeighborsClassifier()
k.fit(X_train,y_train)
```

```python
y_pred=k.predict(X_test)
```

```python
print(k.predict([[1,2,4,1,2,3.2,4.3,2.3,1.5,6.2,2.7]]))
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,f1_score
```

```python
conf_matrix=confusion_matrix(y_test,y_pred)
sns.heatmap(conf_matrix,annot=True)
plt.title('confusion matrix')
plt.xlabel('actual')
plt.ylabel('predicted')
plt.show()
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
