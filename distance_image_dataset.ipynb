```python
import numpy as np
import pandas as pd
```

```python
def euclidean(p1,p2):
    p1=np.array(p1)
    p2=np.array(p2)
    d=np.sqrt(np.sum((p1-p2)**2))
    return d
p1=(4,5)
p2=(3,2)
euclidean(p1,p2)
```

```python
def manhattan(p1,p2):
    p1=np.array(p1)
    p2=np.array(p2)
    d=np.sum(abs(p1-p2))
    return d
p1=(7,6)
p2=(4,5)
manhattan(p1,p2)
```

```python
def lr_norm(p1,p2,r):
    p1=np.array(p1)
    p2=np.array(p2)
    d=(np.sum((abs(p1-p2)**r)))**(1/r)
    return d
p1=(3,5)
p2=(4,7)
r=4
lr_norm(p1,p2,r)
```

```python
A='julie loves me more than linda loves me'
B='jane like me more than julie loves me'
```

```python
def cos_sim(A,B):
    a=[]
    b=[]
    s1=A.split()
    s2=B.split()
    s=s1 or s2
    for i in s:
        if i in s1:
            a.append(int(1))
        else:
            a.append(int(0))
        if i in s2:
            b.append(int(1))
        else:
            b.append(int(0))
    
    a=np.array(a)
    b=np.array(b)
    num=np.sum(a*b)
    den=np.sqrt(np.sum(a**2))*np.sqrt(np.sum(b**2))
    cos=num/den
    return cos

cos_sim(A,B)
    
```

```python
import os
from skimage.transform import resize
from skimage.io import imread
import matplotlib.pyplot as plt
```

```python
categories=['cloudy','desert','green_area','water']
flat_data_arr=[]
target_arr=[]

data=r'C:\Users\22anj\Desktop\images\dataset'
for i in categories:
    count=0
    print(f'loading....categories:{i}')
    path=os.path.join(data,i)
    for img in os.listdir(path):
        if count<200:
            img_array=imread(os.path.join(path,img))
            img_resize=resize(img_array,(150,150,3))
            flat_data_arr.append(img_resize.flatten())
            target_arr.append(categories.index(i))
            
            count+=1
    print(f'loaded categories:{i} success')

    flat_data=np.array(flat_data_arr)
    target=np.array(target_arr)
    
```

```python
df=pd.DataFrame(flat_data)
df['target']=target
df.shape
```
