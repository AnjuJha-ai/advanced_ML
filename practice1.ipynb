```python
from sklearn.feature_selection import mutual_info_regression
```

```python
i=mutual_info_regression(x,y)
f_i=pd.Series(i,df.columns[0:len(df.columns)-1])
f_i.plot(kind='barh',color='red')
plt.show()
```

```python
from sklearn.feature_selection import chi2,SelectKBest
```

```python
selection=SelectKBest(mutual_info_regression,k=3)
```

```python
selection.fit(x,y)
```

```python
print(selection.get_support())
```

```python
X=selection.fit_transform(x,y)
print(x.shape[1])
print(X.shape[1])
```

```python
rank=fisher_score.fisher_score(x,y)
f_i=pd.Series(rank,df.columns[0:len(df.columns)-1])
f_i.plot(kind='barh',color='orange')
plt.show()
```
