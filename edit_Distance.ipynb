```python
import numpy as np
```

```python
S1='INSPECTION'
S2='SUSPENSION'
print(len(S1),len(S2))
```

```python
mat=np.zeros((len(S1)+1,len(S2)+1),dtype=np.uint8)
print(list(S1))
print(list(S2))
```

```python
for i in range(len(mat[0])):
    for j in range(len(mat[1])):
        if i==0:
            mat[i][j]=j
        if i!=0 and j==0:
            mat[i][j]=i
print(mat)
```

```python
import pandas as pd
for i in range(len(S1)):
    for j in range(len(S2)):
        if S1[i]==S2[j]:
            mat[i+1][j+1]=mat[i][j]
        else:
            m=min(mat[i+1][j],mat[i][j],mat[i][j+1])
            mat[i+1][j+1]=m+1
word=np.matrix(mat[1:,1:])
pd.DataFrame(word,index=list(S1),columns=list(S2))
#pd.DataFrame(word)
```
