```python
import numpy as np
import pandas as pd
from sklearn import datasets
```

```python
Wine=datasets.load_wine()
```

```python
x=Wine.data
y=Wine.target
col1=Wine.feature_names
col2=Wine.target_names
```

```python
df=pd.DataFrame(x,columns=col1)
```

```python
df['target'] = y
```

```python
# filter method using information gain of feature for feature selection
from sklearn.feature_selection import mutual_info_classif
```

```python
importances=mutual_info_classif(x,y)
feat_importances=pd.Series(importances,df.columns[0:len(df.columns)-1])
feat_importances.plot(kind='barh',color='Red')
plt.show()
```

```python
# chisquare test is used for categorical features in a dataset.
#calculate chi square between each feature and the target and select the desired number of features with the best chi square score.
# in order to correctly apply chi squared to test the relation between various feature
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
#convert categorical by integer
#x=x.astype(int)
#three features with highest chi-squared statistics are selected
chi2_features=SelectKBest(chi2,k=5)
x_bestfeature=chi2_features.fit_transform(x,y)

#Reduced features

print('Original features:',x.shape[1])
print('reduced feature:',x_bestfeature.shape[1])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,mean_absolute_error,mean_squared_error
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
#elbow method
error_rate=[]
for i in range(1,10):
    model=KNeighborsClassifier(n_neighbors=i)
    model.fit(x_train,y_train)
    prediction=model.predict(x_test)
    error_rate.append(np.mean(prediction!=y_test))
```

```python
plt.figure(figsize=(8,8))
plt.plot(range(1,10),error_rate,color='blue',linestyle='dotted',marker='x')
plt.title('Error rate vs k')
plt.xlabel('k')
plt.ylabel('error rate')
```
