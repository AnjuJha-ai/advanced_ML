```python
# Que2:Implement the model of simple linear regression using the formula of m(slope) & c(intercept).Find R2  and RMSE values.
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
x=np.array([2,4,5,3,8,10]).reshape((-1,1))
y=np.array([7,5,12,8,12,16])
plt.plot(x,y,'*',color="Red")
plt.title('datapoints on graph')
plt.show()
```

```python
mean_x = np.mean(x)
mean_y = np.mean(y)

# Total number of values
m = len(x)

# Using the formula to calculate b1 and b2
num = 0
den = 0
for i in range(m):
    num += (x[i] - mean_x) * (y[i] - mean_y)
    den += (y[i] - mean_x) ** 2
b1 = num / den
b0 = mean_y - (b1 * mean_x)

# Print coefficients
print("Slope:",b1,"intercept:", b0)
```

```python
 #Plotting Values and Regression Line

max_x = np.max(x) + 10
min_x = np.min(x) - 10

# Calculating line values x and y
X = np.linspace(min_x, max_x, 1000)
Y = b0 + b1 * X

# Ploting Line
plt.plot(X, Y, color='blue', label='Regression Line')
# Ploting Scatter Points
plt.scatter(x, y, color='black', label='Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Regression line across datapoints')
plt.show()
```

```python
# Calculating Root Mean Squares Error
rmse = 0
for i in range(m):
    y_pred = b0 + b1 * x[i]
    rmse += (y[i] - y_pred) ** 2
rmse = np.sqrt(rmse/m)
print("RMSE:",rmse)
```

```python
ss_t = 0
ss_r = 0
for i in range(m):
    y_pred = b0 + b1 * x[i]
    ss_t += (y[i] - mean_y) ** 2
    ss_r += (y[i] - y_pred) ** 2
r2 = 1 - (ss_r/ss_t)
print("r2_score:",r2)
```

```python
# Que4: Implement model fitting prediction and score finding for simple linear regression using the sci-kit learn library.
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Cannot use Rank 1 matrix in scikit learn
x = x.reshape((m, 1))
# Creating Model
reg = LinearRegression()
# Fitting training data
reg = reg.fit(x, y)
# Y Prediction
y_pred = reg.predict(x)

# Calculating RMSE and R2 Score
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)
r2_score = reg.score(x, y)

print("MSE:",np.sqrt(mse))
print("r2_score:",r2_score)
```
