```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python

img=cv2.imread('BW.jpg',0)
row,col=img.shape
M1=np.float32([[1,0,100],[0,1,50]])    #translation
dest=cv2.warpAffine(img,M1,(row,col))
plt.imshow(dest,cmap='gray')
```

```python
#reflection
img=cv2.imread('bird.jpg',0)
row,col=img.shape
M2=np.float64([[-1,0,col],[0,1,0],[0,0,1]])
ref_img=cv2.warpPerspective(img,M2,(int(col),int(row)))

M3=np.float32([[1,0,0],[0,-1,row],[0,0,1]])
ref_img1=cv2.warpPerspective(img,M3,(int(col),int(row)))
plt.subplot(2,1,1)
plt.imshow(ref_img,cmap='gray')
plt.subplot(2,1,2)
plt.imshow(ref_img1,cmap='gray')
plt.show()
```

```python
#rotation
img=cv2.imread('bird.jpg',0)
row,col=img.shape
M4=np.float32([[1,0,0],[0,-1,row],[0,0,1]])
img_ro=cv2.warpAffine(img,cv2.getRotationMatrix2D((col/2,row/2),45,0.6),(col,row))
plt.imshow(img_ro,cmap='gray')
                      
```

```python
#scaling
img=cv2.imread('bird.jpg',0)
row,col=img.shape
img_shrink=cv2.resize(img,(500,400),interpolation=cv2.INTER_AREA)
img_enlarge=cv2.resize(img_shrink,None,fx=2.5,fy=1.5,interpolation=cv2.INTER_CUBIC)
plt.subplot(2,3,1)
plt.imshow(img_shrink,cmap='gray')
plt.subplot(2,3,2)
plt.imshow(img_enlarge,cmap='gray')
plt.show()
```

```python
#crop
img=cv2.imread('bird.jpg',0)
crop_img=img[300:400,300:400]
cv2.imwrite('cropped1.jpg',crop_img)
plt.imshow(crop_img,cmap='gray')
```

```python
#shear

img=cv2.imread('ram.png',0)
row,col=img.shape
M=np.float32([[1,0.5,0],[0.5,1,0],[0,0,1]])
img_shear=cv2.warpPerspective(img,M,(int(row*2),int(col*1.5)))
plt.imshow(img_shear,cmap='gray')
```
