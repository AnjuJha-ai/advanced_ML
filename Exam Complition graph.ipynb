```python
import numpy as np
import matplotlib.pyplot as plt

# Define the bounds
a = 30
b = 75

# Generate x values
x_values = np.linspace(a, b, 1000)

# Calculate the PDF
pdf_values = np.ones_like(x_values) / (b - a)

# Plot the distribution
plt.plot(x_values, pdf_values, label="Uniform Distribution")
plt.xlabel("Time (minutes)")
plt.ylabel("Probability Density")
plt.title("Uniform Distribution for Exam Completion Time")
plt.grid(True)
plt.legend()
plt.show()
```
