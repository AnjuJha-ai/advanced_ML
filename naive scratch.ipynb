```python
def prob(d_f,x,cat,clas):
    yes=[]
    yes_class=d_f[d_f[x]==cat][clas].value_counts()[1]
    print(yes_class)
    no_class=d_f[d_f[x]==cat][clas].value_counts()[0]
    print(no_class)
    total_prob=yes_class+no_class
    print(total_prob)

    total_len=len(d_f[clas])
    for i in d_f[clas]:
        if i==1:
            yes.append(i)
            
    prob_cat=(yes_class/sum(yes))
    
    print('yes_sunny given total_yes :',prob_cat)
    
    prob_yes=len(yes)/len(d_f)
    
    print('total_yes given total_obser:',prob_yes)
            
prob(df,'outlook','sunny','play')
prob(df,'outlook','rainy','play')
prob(df,'outlook','overcast','play')

    
```
