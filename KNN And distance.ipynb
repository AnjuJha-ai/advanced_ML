```python
def euclidean_distance(row1,row2):
    distance=0.0
    print(len(row1))
    for i in range(len(row1)-1):
        distance+=(row1[i]-row2[i])**2
    return sqrt(distance)
```

```python
dataset=[[1,2,3],[1.2,2.4,3.2],[2.6,2,3],[1,3,2],[5,4,3],[3,4,5],[6,5,4],[7,4,2]]
row0=dataset[0]
print(row0)
for row in dataset:
    print(row)
    distance=euclidean_distance(row0,row)
    print(distance)
```

```python
def get_neighbour(train,test_row,num_neighbour):
    distances=list()
    for train_row in train:
        dist=euclidean_distance(test_row,train_row)
        distances.append((train_row,dist))
        distances.sort(key=lambda tup: tup[1])
    neighbours=list()
    for i in range(num_neighbour):
        neighbours.append(distances[i][0])
    return neighbours
#get_neighbour(4,4,2)
```

```python
def Haming_distance(row1,row2):
    distance=0.0
    print(len(row1))
    for i in range(len(row1)-1):
        ham_distance+=abs((row1[i]-row2[i]))
    return (ham_distance)
```
