```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from cv2 import warpAffine

def transf(path,k):
    img=cv2.imread(path,0)
    img=cv2.resize(img,(1000,1000))
    w,h=img.shape
    copy_img=np.zeros_like(img,dtype=np.uint8)
    for i in range(w):
        for j in range(h):
            mat=np.array([[i],[j],[1]])
            resu=np.dot(k,mat)
            m=int(resu[0][0])
            n=int(resu[1][0])
            # print(m,n)
            if m<h and n<w:
                copy_img[m][n]=img[i][j]
    plt.imshow(copy_img)
    plt.show()

kernel=np.array([[1,0,500],[0,1,0],[0,0,1]])
path="BW.jpg"
transf(path,kernel)

img = cv2.imread("BW.JPG", 0)
matrix = np.array([[1, 0, 500], [0, 1, 0]], dtype=np.float32)
sata = cv2.warpAffine(img, matrix, (1000, 1000))
plt.imshow(sata)
```
