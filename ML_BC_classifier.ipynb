```python
from sklearn.feature_selection import mutual_info_classif
from sklearn.feature_selection import SelectKBest
```

```python
priority=mutual_info_classif(x,y)
```

```python
feature=pd.Series(priority,df.columns[0:len(df.columns)-1])
feature.plot(kind='barh',color='red')
plt.show()
```

```python
from sklearn.feature_selection import chi2
chi_feature=SelectKBest(chi2,k=7)
best_feature=chi_feature.fit_transform(x,y)
```

```python
print('original feature:',x.shape[1])
print('best feature:',best_feature.shape[1])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x=df.drop('target',axis=1)
x
```

```python
y=df['target']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x.shape
```

```python
sns.distplot(x)
```

```python
sns.distplot(y)
```

```python
l=LogisticRegression()
l.fit(x_train, y_train)
```

```python
y_pred=l.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,f1_score
```

```python
b=confusion_matrix=(y_test,y_pred)
sns.heatmap(b,annot=True)
```

```python
plt.plot(b)
```

```python
from sklearn.metrics import classification_report
print(classification_report,y_test,y_pred)
accuracy_score(y_test,y_pred)
```
