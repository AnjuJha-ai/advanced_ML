```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
diabetes=datasets.load_diabetes()
```

```python
x=diabetes.data
y=diabetes.target
```

```python
col1=diabetes.feature_names
col2=diabetes.target_filename
```

```python
df=pd.DataFrame(x,columns=col1)
```

```python
df.head()
```

```python
df['target']=y
```

```python
df.head()
```

```python
import seaborn as sns
plt.figure(figsize=(20,5))
sns.heatmap(df.corr(),annot=True)
```

```python
df.info()
```

```python
x=df.drop('target',axis=1)
```

```python
x
```

```python
y=df['target']
```

```python
sns.boxplot(x)
```

```python
x.clip(x.quantile(0.25),x.quantile(0.75),axis=1,inplace=True)
```

```python
sns.boxplot(x)
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler,Normalizer
```

```python
s=StandardScaler()
x_Stand=s.fit_transform(x)
```

```python
x1=pd.DataFrame(x_Stand,columns=col1)
x1.head()
```

```python
plt.scatter(x1['bmi'],y)
```

```python
plt.scatter(x1['bp'],y)
```

```python
from sklearn.feature_selection import SelectKBest
```

```python
from sklearn.feature_selection import mutual_info_regression
```

```python
a=SelectKBest(mutual_info_regression,k=5)
```

```python
X=a.fit_transform(x1,y)
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.40,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
l=LinearRegression()
l.fit(X_train,y_train)
```

```python
y_pred=l.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
r2=r2_score(y_test,y_pred)
print('Root mean square:',r2)
```

```python
MSE=mean_squared_error(y_test,y_pred)
print('MSE:',MSE)
```

```python
MAE=mean_absolute_error(y_test,y_pred)
print('MAE:',MAE)
```

```python
plt.scatter(y_test,y_pred)
```

```python
print(f'b1 regression coefficient:{l.intercept_}')
```

```python
print(f'b0 constant:{l.coef_}')
```
