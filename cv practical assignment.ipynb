```python
'''7. Perform transformations on image using hardcoding with matrix multiplication of
image and matrices of transformation'''
import cv2
import matplotlib.pyplot as plt
import numpy as np
image=cv2.imread('sobel.png',0)
# Define transformation matrices
# Translation matrix
translation_matrix = np.float32([[1, 0, 100],    # x-translation
                                 [0, 1, 50]])    # y-translation

# Rotation matrix (around the center of the image)
angle = 45  # in degrees
center_x = image.shape[1] // 2
center_y = image.shape[0] // 2
rotation_matrix = cv2.getRotationMatrix2D((center_x, center_y), angle, 1)  # 1 is the scaling factor

# Scaling matrix
scaling_matrix = np.float32([[0.5, 0, 0],   # Scale in x-direction
                              [0, 0.5, 0]])  # Scale in y-direction

# Apply transformations using matrix multiplication
translated_image = cv2.warpAffine(image, translation_matrix, (image.shape[1], image.shape[0]))
rotated_image = cv2.warpAffine(image, rotation_matrix, (image.shape[1], image.shape[0]))
scaled_image = cv2.warpAffine(image, scaling_matrix, (image.shape[1], image.shape[0]))

# Display the original and transformed images
cv2.imshow('Original Image', image)
cv2.imshow('Translated Image', translated_image)
cv2.imshow('Rotated Image', rotated_image)
cv2.imshow('Scaled Image', scaled_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
'''9. Perform Segmentation using
- Simple thresholding
- Otsu Binarization
- Adaptive Thresholding'''
image=cv2.imread('ram.png',0)
# Simple Thresholding
_, simple_threshold = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)

# Otsu's Binarization
_, otsu_threshold = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Adaptive Thresholding
adaptive_threshold = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)

# Display the results
cv2.imshow('Original Image', image)
cv2.imshow('Simple Thresholding', simple_threshold)
cv2.imshow('Otsu Binarization', otsu_threshold)
cv2.imshow('Adaptive Thresholding', adaptive_threshold)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
