```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('SOBEL.png',cv2.IMREAD_GRAYSCALE)
kernel=np.ones((5,5),np.uint8)
e_img=cv2.erode(img,kernel,iterations=6)

plt.imshow(e_img,cmap='gray')
```

```python
plt.imshow(img,cmap='gray')
```

```python
kernel1=np.ones((5,5),np.uint8)
d_img=cv2.dilate(e_img,kernel1,iterations=5)
plt.imshow(d_img,cmap='gray')
```

```python
b=img-e_img
plt.imshow(b,cmap='gray')
```

```python
c=img-d_img
plt.imshow(c,cmap='gray')
```
