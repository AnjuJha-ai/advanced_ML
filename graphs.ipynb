# Graphical representation for p/(1-p)

```python


import matplotlib.pyplot as plt
import numpy as np

# Generate values for p (probability) between 0 and 1
x = np.linspace(0.01, 0.99,100)  # sample number to Avoid division by zero

# Calculate the corresponding y-values for p / (1 - p)
y =x/(1 - x)

# Create the plot
# plt.figure(figsize=(5, 5))
plt.plot(x, y, label='p/(1 - p)', color='red')
plt.xlabel('p (Probability)')
plt.ylabel('p/(1 - p)')
plt.title('Graph of p / (1 - p)')
# plt.grid(True)
#plt.legend()
plt.show()
```

# Graphical representation for log(p/(1-p))

```python
import math as m
x = np.linspace(0.01, 0.99,100)  # sample number to Avoid division by zero

# Calculate the corresponding y-values for p / (1 - p)
y =np.log(x/(1 - x))

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='log(p/(1 - p))', color='red')
plt.xlabel('p (Probability)')
plt.ylabel('log(p/(1 - p))')
plt.title('Graph of log(p / (1 - p))')
plt.grid(True)
#plt.legend()
plt.show()
```

# Graphical representation of sigmoid function(z)=1/(1+e^(-z))

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1.0 / (1.0 + np.exp(-x))

# Generate x values from -10 to 10
#x = np.linspace(-100, 100, 10)
x=np.arange(-10,10,0.1)

# Plot the sigmoid function
plt.plot(x, sigmoid(x),color='red')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.title('Sigmoid Function')
plt.grid()
plt.show()
```
