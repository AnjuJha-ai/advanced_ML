```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn import datasets
```

```python
bc=datasets.load_breast_cancer()
x=bc.data
y=bc.target
```

```python
col=bc.feature_names
```

```python
df=pd.DataFrame(x,columns=col)
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
df.describe()
```

```python
plt.figure(figsize=(15,6))
sns.boxplot(df)
```

```python
df.isnull().sum()
```

```python
df.duplicated().value_counts()
```

```python
df.clip(df.quantile(0.25),df.quantile(0.75),axis=1,inplace=True)
```

```python
sns.boxplot(df)
```

```python
from sklearn.preprocessing import StandardScaler,Normalizer,MinMaxScaler,RobustScaler
```

```python
S=StandardScaler()
```

```python
X=S.fit_transform(df)
```

```python
X=pd.DataFrame(X,columns=col)
```

```python
X.head()
```

```python
df['target']=y
```

```python
df
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),annot=True)
```

```python
from sklearn.feature_selection import SelectKBest,mutual_info_classif
```

```python
i=mutual_info_classif(X,y)
feat=pd.Series(i,df.columns[0:len(df.columns)-1])
feat.plot(kind='barh',color='red')
plt.show()
```

```python
a=SelectKBest(mutual_info_classif,k=10)
R=a.fit_transform(X,y)
```

```python
R.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
R_train, R_test, y_train, y_test = train_test_split(R, y, test_size=0.33, random_state=42)
```

```python
k=KNeighborsClassifier()
```

```python
k.fit(R_train,y_train)
```

```python
y_pred=k.predict(R_test)
```

```python
print(k.predict([[2,5,9,3.8,5.7,124,236,21,57,900]]))
```

```python
from sklearn.metrics import confusion_matrix,classification_report,f1_score,accuracy_score
```

```python
b=confusion_matrix(y_pred,y_test)
sns.heatmap(b,annot=True)
plt.xlabel('prediction')
plt.ylabel('actual')
plt.show()
```

```python
c=accuracy_score(y_test,y_pred)
print(c)
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.svm import SVC
```

```python
D=SVC()
```

```python
D.fit(R_train,y_train)
```

```python
y_p=D.predict(R_test)
```

```python
print(accuracy_score(y_p,y_test))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
p={'C':[0.01,0.1,1,0.001],'kernel':['linear','poly','rbf','sigmoid'],'gamma':['scale','auto'],'degree':[1,2,3,4,5]}
```

```python
g=GridSearchCV(D,p,cv=3)
```

```python
g.fit(R_train,y_train)
```

```python
y_pr=g.predict(R_test)
```

```python
g.best_params_
```

```python
A=accuracy_score(y_pr,y_test)
print(A)
```

```python
D=SVC(C=100,gamma='auto',kernel='poly',degree=3)
```

```python
D.fit(R_train,y_train)
```

```python
y1=D.predict(R_test)
```

```python
q=accuracy_score(y1,y_test)
print(q)
```

```python
D=SVC(C=0.01,gamma='scale',kernel='rbf',degree=3)
```

```python
D.fit(R_train,y_train)
```

```python
y2=D.predict(R_test)
```

```python
w=accuracy_score(y2,y_test)
print(w)
```

```python
name=['poly','linear','rbf']
accuracy=[q,A,w]
plt.bar(name,accuracy)
plt.xlabel('function')
plt.ylabel('accuracy')
plt.show
```
