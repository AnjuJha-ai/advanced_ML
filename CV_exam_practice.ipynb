```python
def multiply_const(img,constant):
    r1=cv2.multiply(img,constant)
    plt.imshow(r1)
# multiply_const(img,4)


def divide_const(img,constant):
    r2=cv2.divide(img,constant)
    plt.imshow(r2)
divide_const(img,5)

def resize(img,scale):
    w=int(img.shape[1]*scale/100)
    h=int(img.shape[0]*scale/100)
    r3=cv2.resize(img,(w,h),interpolation=cv2.INTER_AREA)
    plt.imshow(r3)
resize(img,45)

def slice_img(img,x,y,w,h):
    w=img.shape[1]
    h=img.shape[0]
    r4=img[y:y+h,x:x+w]
    plt.imshow(r4)
slice_img(img,100,100,200,200)

def mask_img(img,mask):
    mask=np.zeros(img.shape[:2], dtype="uint8")
    center = (img.shape[1]//2, img.shape[0]//2)
    radius = 100
    cv2.circle(mask,center,radius,200,-1)
       
    mask_img=cv2.bitwise_and(img,img,mask=mask)
    plt.imshow(mask_img)
mask=np.zeros(img.shape[:2], dtype="uint8")
mask_img(img,mask)

def add_img(img,img1):
    
    r5=cv2.add(img[0:200,0:200],img1[0:200,0:200])
    plt.imshow(r5)
img1=cv2.imread('BW.jpg') 
add_img(img,img1)

def weight_add(img,img1,alpha,beta,gamma):
    r6=cv2.addWeighted(img,alpha,img1,beta,gamma)
    plt.imshow(r6)
weight_add(img,img1,0.6,0.4,0)

def sub_img(img,img1):
    r7=cv2.subtract(img,img1)
    plt.imshow(r7)
sub_img(img,img1)

def logical_op(img):
    mask=cv2.inRange(img,(100,100,100),(255,255,255))
    
    bitwise_and=cv2.bitwise_and(img,img1,mask=mask)
    bitwise_or=cv2.bitwise_or(img,img1,mask=mask)
    bitwise_not=cv2.bitwise_not(img,img1,mask=mask)
    plt.imshow(bitwise_and)
    plt.imshow(bitwise_or)
    plt.imshow(bitwise_not)
logical_op(img)

```

```python
def img_distance(p1,p2):
    d=0
    for i in range(len(p1)):
        d+=(p1[i]-p2[i])**2
    print('distance:',np.sqrt(d))
p1=(4,5)
p2=(7,3)
img_distance(p1,p2)
        
```

```python
def neg(img):
    imgw=255-img
    plt.subplot(321)
    plt.imshow(img)
    plt.subplot(322)
    plt.imshow(imgw)
neg(img)
```

```python
def log_trans(img):
    c=255/np.log(1+np.max(img))
    log_img=c*np.log(1+img)
    log_img=np.array(log_img,dtype=np.uint16)
    log_img=cv2.cvtColor(log_img,cv2.COLOR_RGB2BGR)
    plt.subplot(321)
    plt.imshow(img,cmap='gray')
    plt.subplot(322)
    plt.imshow(log_img)
img=cv2.imread('face.jpg',0)
log_trans(img)
```

```python
def power_log(img,p):
    c=int(input('enter value:'))
    p_log=np.power(c*np.log(1+img),p)
    p_log=np.array(p_log,dtype=np.uint16)
    p_log=cv2.cvtColor(p_log,cv2.COLOR_RGB2BGR)
    plt.imshow(p_log,cmap='gray')
power_log(img,3)
```

```python
#gray level
img=cv2.imread('bird.jpg',0)
def gray_level(img,max_i,min_i):
    img1=img.copy()
    img1[(img>=min_i) & (img<=max_i)]=255
    img1[(img<min_i)|(img>max_i)]=0
    plt.imshow(img1,cmap='gray')
gray_level(img,200,50)
```

```python
# hist, bins = np.histogram(img.flatten(), 255, [0, 255])
hist,bins=np.histogram(img.flatten(),255,[0,255])
c=hist.cumsum()
c_norm=c*hist.max()/c.max()
boxplot_slice=np.interp(img.flatten(),bins[:-1],c_norm)
boxplot_slice=boxplot_slice.reshape(img.shape)

cv2.imshow('Boxplot Sliced Image', boxplot_slice.astype(np.uint8))
cv2.waitKey(0)
```
