```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the normal distribution
mu = 0
sigma = 1

# Generate data points for the normal distribution(pdf:probability density function)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 1000)
pdf_normal = norm.pdf(x, mu, sigma)

# Generate data points for the standard normal distribution (Z)
z = np.linspace(-3, 3, 1000)
pdf_standard_normal = norm.pdf(z, 0, 1)

# Create subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Plot the normal distribution
ax1.plot(x, pdf_normal, label='Normal Distribution')
ax1.set_title('Normal Distribution')
ax1.set_xlabel('x')
ax1.set_ylabel('Probability Density')
ax1.grid(True)

# Plot the standard normal distribution (Z)
ax2.plot(z, pdf_standard_normal, label='Standard Normal Distribution')
ax2.set_title('Standard Normal Distribution (Z)')
ax2.set_xlabel('z')
ax2.set_ylabel('Probability Density')
ax2.grid(True)

# Shade the area under the standard normal distribution curve up to Z = 1.96
z_value = 1.96
ax2.fill_between(z, pdf_standard_normal, where=(z <= z_value), alpha=0.3, color='blue', label=f'Z ≤ {z_value:.2f}')

# Add legend
ax1.legend()
ax2.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
def main():
    # Input the sample space of events A and B
    sample_A = input("Enter the sample for event A: ").split(",")
    sample_B = input("Enter the sample for event B: ").split(",")

    # Convert sample spaces to sets
    set_A = set(sample_A)
    set_B = set(sample_B)

    # Intersection of sets A and B
    intersection_AB = set_A & set_B

    # Calculate conditional probabilities
    P_A_given_B = len(intersection_AB) / len(set_B)
    P_B_given_A = len(intersection_AB) / len(set_A)

    # Check if events are mutually exclusive
    mutually_exclusive = len(intersection_AB) == 0

    # Check if events are independent
    independent = P_A_given_B == P_B_given_A

    # Print results
    print(f"P(A|B) = {P_A_given_B:.4f}")
    print(f"P(B|A) = {P_B_given_A:.4f}")
    print(f"Mutually Exclusive: {mutually_exclusive}")
    print(f"Independent: {independent}")

if __name__ == "__main__":
    main()
```
